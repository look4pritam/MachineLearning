```python
import nemo.collections.asr as nemo_asr
speaker_model = nemo_asr.models.EncDecSpeakerLabelModel.from_pretrained("nvidia/speakerverification_en_titanet_large")
```

```python
embedding = speaker_model.get_embedding("an255-fash-b.wav")
```

```python
print(embedding.shape)
```

```python
speaker_model.verify_speakers("an255-fash-b.wav","cen7-fash-b.wav")
```

```python
import torch

def compare_samples(path1, path2):
    if not (path1 and path2):
        return '<b style="color:red">ERROR: Please record audio for *both* speakers!</b>'

    embs1 = speaker_model.get_embedding(path1).squeeze()
    embs2 = speaker_model.get_embedding(path2).squeeze()
    
    #Length Normalize
    X = embs1 / torch.linalg.norm(embs1)
    Y = embs2 / torch.linalg.norm(embs2)
    
    # Score
    similarity_score = torch.dot(X, Y) / ((torch.dot(X, X) * torch.dot(Y, Y)) ** 0.5)
    print('similarity_score', similarity_score)
    similarity_score = (similarity_score + 1) / 2

    print('similarity_score', similarity_score)
```

```python
compare_samples("an255-fash-b.wav","cen7-fash-b.wav")
```
